# Geometric Random Walk

This simulation generates 1000 days of random prices based on a starting price and a daily range

You can play around with the starting_price on Day 0 and daily_range that tells how much the price can possibly move in a day.

```python
import random
import pandas as pd
import plotly.express as px

# Starting price of scrip
starting_price = 100

# % change per day (lower_range,higher_range)
daily_range = (-1,1.1)

class PriceGenerator:
    def __init__(self,starting_price,daily_range):
        self.current_price = starting_price
        self.daily_range = daily_range

    def get(self) -> float:
        self.current_price = self.current_price * (1 + random.uniform(*self.daily_range) * 0.01)
        return self.current_price

simulator = PriceGenerator(starting_price,daily_range)
prices = [simulator.get() for _ in range(0,1000)]

px.line(x=range(0,1000),y=prices,labels={'x':'Day #','y':'Price'})
```
